```python
marker2 = Pose2D(0.35, 0.6, 0)
marker3 = Pose2D(0.7, 0, 0)
pose_estimator = TripleMarkerPoseEstimator(marker2, marker3)
```

```python
pose_estimator.compute_pose(np.pi/10, np.pi/20, -np.pi/40)
print(pose_estimator)
```

```python
x2, y2 = marker2.x, marker2.y
x3, y3 = marker3.x, marker3.y
x, y, t = pose_estimator.offset_x, pose_estimator.offset_y, pose_estimator.rotation
plt.figure(1, figsize=[10, 10])
plt.scatter(0, 0, label="Marker 1")
plt.scatter(x2, y2, label="Marker 2")
plt.scatter(x3, y3, label="Marker 3")
plt.scatter(x, y, label="Camera Pos")
plt.plot([x, 0], [y, 0])
plt.plot([x, x2], [y, y2])
plt.plot([x, x3], [y, y3])
plt.plot([x, x + np.cos(t)], [y, y + np.sin(t)], label="Normal Line of Cam")
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.legend()
```

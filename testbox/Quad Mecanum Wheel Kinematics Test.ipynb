```python
import numpy as np
```

```python
def wheel_vel(vx, vy, omega):
    a = np.array([ np.pi/4, 3*np.pi/4, 5*np.pi/4, 7*np.pi/4 ])    # Alpha
    b = np.array([ np.pi/4, -np.pi/4, -3*np.pi/4, 3*np.pi/4 ])    # Beta
    g = np.array([ np.pi/4, -np.pi/4, -np.pi/4, np.pi/4 ])   # Gamma
    L = np.array([ np.sqrt(2), np.sqrt(2), np.sqrt(2), np.sqrt(2) ])
    R = 0.05
    
    return (-vx - vy * np.tan(a + b + g) - L * omega * np.sin(b + g) / np.cos(a + b + g)) / (R * np.sin(g) / np.cos(a + b + g))
```

```python
wheel_vel(1, 0, 0)
```

```python
wheel_vel(np.pi/4, np.pi/4, 1)
```

```python
t = np.linspace(0, 2 * np.pi, 100)
vx = np.cos(t)
vy = np.sin(t)
```

```python
vel = np.zeros([100, 4])
for x in range(100):
    vel[x,:] = wheel_vel(vx[x], vy[x], 1)
#     vel[x,:] = wheel_vel(0, 0, t[x])
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(1, figsize=[15,7])
for x in range(4):
    plt.subplot(2, 2, x+1)
    plt.plot(t, vel[:,x])
    plt.title("Wheel %d" % (x+1))
```
